# Migrating Data to SQLite DataBase

```python
# Set up connection
conn=sqlite3.connect('Resources/db.sqlite')
```

## Migrating Climate Change Data

```python
# Convert Pandas DataFrame to SQL Database
climatechange_df.to_sql('climateChange_table', conn, if_exists='replace', index=False)
```

```python
#  Confirm data is in SQL database.
pd.read_sql('select * from climateChange_table', conn)
```
