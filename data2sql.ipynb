# Migrating Data to SQLite DataBase

```python
# Set up connection
conn=sqlite3.connect('Resources/db.sqlite')
```

## Migrating Sealevel Data

```python
# Convert Pandas DataFrame to SQL Database
sealevel_df.to_sql('sealevel_db', conn, if_exists='replace', index=False)
```

```python
#  Confirm data is in SQL database.
pd.read_sql('select * from sealevel_db', conn)
```
